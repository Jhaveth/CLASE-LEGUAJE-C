#EJEMPLO 1: Hacer un programa para rellenar una matriz pidiendo al usuario el número de filas y columnas, Posteriormente mostrar la matriz en pantalla

```python
%%writefile ejemplo1.cpp

#include<iostream>
//Aquí deberas tambien agregar la funcion #include<conio.h>
using namespace std;
int main() {
    int filas, columnas;
    cout << "Ingrese el numero de filas: ";
    cin >> filas;
    cout << "Ingrese el numero de columnas: ";
    cin >> columnas;
    int matriz[filas][columnas];
    for (int i = 0; i < filas; i++) {
        for (int j = 0; j < columnas; j++) {
            cout << "Ingrese el valor para la posicion [" << i << "][" << j << "]: ";
            cin >> matriz[i][j];
        }
    }
    cout << "La matriz ingresada es:" << endl;
    for (int i = 0; i < filas; i++) {
        for (int j = 0; j < columnas; j++) {
            cout << matriz[i][j] << " ";
        }
        cout << endl;
    }
    return 0;
    //getch(); se deberá usar en c++
}
```

```python
# Paso 3: Compilar el código
%%shell
g++ ejemplo1.cpp -o mi_programa.cpp
```

#EJEMPLO 2: Hacer una matriz de tipo entera de 2*2, llenarla de numeros y luego copiar todo su contenido hacia otra matriz.

```python
%%writefile ejemplo2.cpp

#include<iostream>
//Aquí deberas tambien agregar la funcion #include<conio.h>
using namespace std;
int main() {
    int matriz[2][2]={{1,2},{3,4}};
    int matriz2[2][2];
    for (int i = 0; i < 2; i++) {
        for (int j = 0; j < 2; j++) {
            matriz2[i][j]=matriz[i][j];
        }
    }
    cout << "La matriz ingresada es:" << endl;
    for (int i = 0; i < 2; i++) {
        for (int j = 0; j < 2; j++) {
            cout << matriz[i][j] << " ";
        }
        cout << endl;
    }
    cout << "La matriz nueva es:" << endl;
    for (int i = 0; i < 2; i++) {
        for (int j = 0; j < 2; j++) {
            cout << matriz2[i][j] << " ";
        }
        cout << endl;
    }

    return 0;
    //getch(); se deberá usar en c++
}
```

#EJEMPLO 3: Realice un programa que lea una matriz de 3x3 y cree su matriz trnaspuesta. Una matriz transpuesta es aquella en la que la columna i era la fila i de la matriz original.

```python
%%writefile ejemplo3.cpp

#include<iostream>
//Aquí deberas tambien agregar la funcion #include<conio.h>
using namespace std;
int main() {
    int matriz[3][3];

    for (int i = 0; i < 3; i++) {
        for (int j = 0; j < 3; j++) {
            cout << "Ingrese el valor para la posicion [" << i << "][" << j << "]: ";
            cin >> matriz[i][j];
        }
    }
    cout << "La matriz original es:" << endl;
    for (int i = 0; i < 3; i++) {
        for (int j = 0; j < 3; j++) {
            cout << matriz[i][j] << " ";
        }
        cout << endl;
    }
    cout << "La matriz transpuesta es:" << endl;
    for (int i = 0; i < 3; i++) {
        for (int j = 0; j < 3; j++) {
            cout << matriz[j][i] << " ";
        }
        cout << endl;
    }


    return 0;
    //getch(); se deberá usar en c++
}
```
